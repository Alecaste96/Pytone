Ecco il primo esempio di rete neurale. 
Ciò mi è servito a comprendere, facendo una semplice rete neurale, la struttura delle stesse e la gestione dei dati in INPUT/OUT

```python
from keras import*
from keras import Sequential
from keras import layers

model=Sequential()

#input costituito da un unita
model.add(layers.Dense(units=3,input_shape=[1]))

#strato intermedio
model.add(layers.Dense(units=2))

#unita strato di uscita sono pari a 1

model.add(layers.Dense(units=1))

rete_input = np.array([1, 2, 3, 4, 5])
rete_output = np.array([3, 6, 9, 12, 15])

#creazione della rete (adam)

model.compile(loss='mean_squared_error',optimizer='adam')

#si allena con 1000 cicli, non mi da risultari perfetti ma con 6000 cicli si ottiene il risultato atteso
model.fit(x=rete_input,y=rete_output,epochs=1000)

print(model.predict(np.array([6])))  # Converti la lista in array NumPy
print(model.predict(np.array([7])))  # Converti la lista in array NumPy
print(model.predict(np.array([8])))  # Converti la lista in array NumPy
```

ORA ho utilizzato lo schema della rete neurale precedente per fare un esempio in ambito calcistico.
l idea di fondo è quella di, a partire da numero di goal,assist,passaggi e tackel fatti da un numero finito di centrocampisti,
andare a predirre il numero di goal e assist che un centrocampista fara nella stagione calcistica futura

```python
#Spiegazione del Codice:
#Input Layer:
#Ha 4 unità, ognuna delle quali rappresenta un tipo di dato storico: goal, assist, passaggi completati e tackle fatti.
#Hidden Layer:
#Il modello ha uno strato nascosto con 8 unità, usando l'attivazione relu per introdurre non-linearità.
#Output Layer:
#Ha 2 unità: una per il numero previsto di goal e una per il numero previsto di assist nella prossima stagione.
#Compilazione:
#Il modello è compilato con la funzione di perdita mean_squared_error e l'ottimizzatore adam, che è una scelta standard per problemi di regressione.
#Addestramento:

#Il modello viene addestrato con dati di esempio.
#Predizione:

#Il modello fa una previsione sui goal e assist futuri di un nuovo giocatore basandosi sui suoi dati storici.
#Raccolta dei Dati:
#Per utilizzare questo modello in modo efficace, è essenziale avere un dataset di alta qualità con i dati storici dei calciatori, comprendente le statistiche delle ultime stagioni. Maggiori sono i dati che riesci a raccogliere e più accurata sarà la previsione.

#Miglioramenti Possibili:
#Includere più caratteristiche: Potresti aggiungere altre statistiche rilevanti come minuti giocati, numero di tiri, dribbling completati, ecc.
#Tuning dell'architettura del modello: Potresti sperimentare con un numero diverso di layer o unità per migliorare la capacità predittiva del modello.
#Aggiunta di regolarizzazione: Per prevenire l'overfitting, puoi aggiungere dropout o penalità L2 sui pesi del modello.
#Questa rete neurale ti permetterà di stimare le performance future di un calciatore sulla base delle sue performance passate, offrendo uno strumento utile per analizzare i talenti e fare previsioni strategiche.
```

```python
import numpy as np
from keras import Sequential
from keras import layers

# Definizione del modello
model = Sequential()

# Strato d'ingresso (16 unità: 4 caratteristiche per 4 anni)
model.add(layers.Dense(units=64, input_shape=[16], activation='relu'))

# Strato nascosto
model.add(layers.Dense(units=32, activation='relu'))

# Strato di uscita (2 unità: una per goal, una per assist)
model.add(layers.Dense(units=2))

# Compilazione del modello
model.compile(loss='mean_squared_error', optimizer='adam')

# Dati di esempio (simulazione) per 4 giocatori con 4 anni di dati
# Input: [goal_anno1, assist_anno1, passaggi_anno1, tackle_anno1, ..., tackle_anno4]
input_data = np.array([
    [1,4,183,36,3,7,261,58,3,11,231,63,6,6,214,49],  # Giocatore 1 Barella
    [1,0,190,67,0,2,82,39,2,1,140,44,2,2,132,59],  # Giocatore 2 Bennacer
    [1,0,58,37,0,0,43,7,1,2,97,36,0,1,83,25],  # Giocatore 3 Cataldi
    [1,3,137,25,1,1,85,40,2,0,231,79,1,1,170,33],  # Giocatore 4 Cristante
])

# Output: [goal_futuro, assist_futuro]
output_data = np.array([
    [2, 3],  # Previsione goal e assist Giocatore 1 Barella
    [2, 2], # Previsione goal e assist Giocatore 2 Bennacer
    [1, 0],   # Previsione goal e assist Giocatore 3 Cataldi
    [3, 4], # Previsione goal e assist Giocatore 4 Cristante
])

# Addestramento del modello
model.fit(x=input_data, y=output_data, epochs=6000)

# Previsione per un nuovo giocatore
new_player_data = np.array([[0,5,229,76,4,2,315,81,3,3,147,44,0,2,134,42]])
prediction = model.predict(new_player_data)

print("Previsione goal e assist per il nuovo giocatore:", prediction)
```

Parlando dei dati utilizzati veri e propri, abbiamo considerato numero di goal,assist,passaggi e tiri di 4 giocatori
(Barella,Bennacer,Cataldi e Cristante) e ci hanno permesso di prevedere numero di goal e assist che fara il quinto giocatore
(Locatelli) nella prossima stagione.

Ora inizia la rete neurale. Viene chiesto il nome del calciatore che si vuole predirre, la rete utilizza tutti i dati
presenti nella tabella e come output ci dara numero di goal e assist del calciatore scelto nel prossimo anno.

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler

# Carica i dati
combined_data = pd.read_excel(r"C:\Users\Mario\Desktop\combined_data.xlsx")  # Modifica il percorso del file se necessario

# Verifica la struttura dei dati

#print(combined_data.head())
#print(combined_data.columns)

# Fa vedere la lista dei giocatori
print("Giocatori presenti in tutti e 5 gli anni: 'Alberto Grassi', 'Ismaël Bennacer', 'Lorenzo Pellegrini', "
      "'Stanislav Lobotka', 'Luis Alberto', 'Nicolò Barella', 'Adrien Rabiot', 'Bryan Cristante', 'Matías Vecino', "
      "'Marten de Roon', 'Roberto Gagliardini', 'Manuel Locatelli', 'Diego Demme', 'Piotr Zieliński', 'Alfred Duncan', "
      "'Mehdi Bourabia', 'Nicolás Domínguez', 'Walace', 'Nicolò Rovella', 'Rolando Mandragora', 'Danilo Cataldi'")

# Chiedi all'utente il nome del giocatore da escludere
player_name = input("Inserisci il nome del giocatore per il quale vuoi fare la previsione: ")

# Trova l'indice del giocatore selezionato
player_index = combined_data[combined_data['Player'] == player_name].index[0]

# Prepara le colonne di input e output
input_columns = combined_data.columns[3:]  # le prime 3 colonne non servono per gli input
output_columns = combined_data.columns[1:3]  # colonna relativa a goal e assist dell'anno che verra predetto

# Estrai i dati di input e output
X = combined_data[input_columns].values
y = combined_data[output_columns].values

# Dividi i dati in set di addestramento e test
X_train = np.delete(X, player_index, axis=0)  # Tutti tranne il giocatore selezionato
y_train = np.delete(y, player_index, axis=0)  # Tutti tranne il giocatore selezionato
X_test = X[player_index:player_index + 1]     # Solo il giocatore selezionato
y_test = y[player_index:player_index + 1]     # Solo il giocatore selezionato

# Normalizzazione dei dati
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)
y_train = scaler_y.fit_transform(y_train)

# Non è necessario scalare y_test per la previsione, ma lo facciamo per completezza
y_test_scaled = scaler_y.transform(y_test)  

# Definizione del modello
model = Sequential()
model.add(Dense(units=64, input_shape=[X_train.shape[1]], activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=2))  # Output: Goal e Assist

# Compilazione del modello
model.compile(loss='mean_squared_error', optimizer='adam')

# Addestramento del modello
model.fit(X_train, y_train, epochs=6000, verbose=1, validation_split=0.1)

# Previsione per il giocatore selezionato
y_pred = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred)  # Riporta i dati alle loro scale originali

print(f"Previsione Goal e Assist per {player_name}: {y_pred}")
```
